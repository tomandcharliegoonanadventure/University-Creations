# Clustr analysis(K-mean)


Cluster analysis is an unsupervised machine learning task. Unlike supervised learning (predictive modeling), clustering algorithms only interpret input data and find groups or clusters in feature Spaces.

In this part, we will use K-means algorithms to classify stocks to different classes according to their features.

```python
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```
