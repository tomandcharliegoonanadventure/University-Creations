```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send an HTTP GET request to the Billboard website
url = "https://www.billboard.com/articles/business/financial-report/9657546/top-earning-musicians-2020"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find the table containing the financial data
table = soup.find("table", class_="inline-table")

# Extract data from the table
rows = table.find_all("tr")
data = []
for row in rows[1:]:  # Exclude the table header row
    columns = row.find_all("td")
    date = columns[0].text.strip()
    artist = columns[1].text.strip()
    earnings = columns[2].text.strip()
    data.append((date, artist, earnings))

# Create pandas dataframe
df = pd.DataFrame(data, columns=["Date", "Artist", "Earnings"])

# Print the dataframe
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_movies_by_year(year):
    # Create the URL for the specific year
    url = f'https://www.imdb.com/search/title/?release_date={year}&sort=num_votes,desc'

    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the movie list
    movie_list = soup.find_all('div', class_='lister-item-content')

    # Create a list of dictionaries to store movie information
    movies = []

    # Iterate over each movie item
    for movie in movie_list:
        # Extract relevant information
        title = movie.find('h3', class_='lister-item-header').find('a').text
        rating = movie.find('div', class_='inline-block ratings-imdb-rating').strong.text

        # Find the movie page URL
        movie_url = 'https://www.imdb.com' + movie.find('h3', class_='lister-item-header').find('a')['href']
        movie_response = requests.get(movie_url)
        movie_soup = BeautifulSoup(movie_response.content, 'html.parser')

        # Find the soundtrack section
        soundtrack_section = movie_soup.find('div', id='soundtracks_content')

        # Extract the soundtracks
        soundtracks_in_movie = []
        if soundtrack_section:
            soundtrack_list = soundtrack_section.find_all('li')
            for soundtrack in soundtrack_list:
                soundtrack_title = soundtrack.text.strip()
                soundtracks_in_movie.append(soundtrack_title)

        # Create a dictionary with the movie information
        movie_data = {
            'Title': title,
            'Rating': rating,
            'Soundtracks': soundtracks_in_movie
        }

        # Append the movie dictionary to the list
        movies.append(movie_data)

    # Create a DataFrame from the list of movie dictionaries
    df = pd.DataFrame(movies)

    return df

# Specify the year for scraping data
year = 2020

# Scrape movies by the specified year and store in a DataFrame
movie_df = scrape_movies_by_year(year)

# Print the DataFrame
print(movie_df)
```
